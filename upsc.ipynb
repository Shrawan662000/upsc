```python
prompt_04_a=f'''
You are tasked with generating UPSC Civil Services Prelims Examination standard questions. Your objective is to create high-quality questions by identifying critical topics and key concepts relevant to the UPSC syllabus from the provided content. Additionally, if the content hints at significant areas but lacks direct details, use your understanding of related topics to frame complex and insightful questions. Ensure the questions align with the analytical rigor and pattern of previous UPSC Prelims exams.

Instructions for Question Generation:
Identify Important Topics and Concepts:

Focus on relevant, high-impact topics in the content that are aligned with UPSC Prelims.
If the content is insufficiently detailed but references significant concepts, use your expertise to expand and frame questions on those concepts.
Generate UPSC Standard Questions:

Avoid simple, factual questions.
Ensure all questions require analysis, interpretation, or application of knowledge.
Diverse Question Types:
Frame a mix of the following question formats to match UPSC standards:

Statement Analysis: "Which of the following statements is/are correct?"
Assertion-Reason: Analyze the relationship between a given assertion and its reason.
Match the Following: Pair complex events, concepts, or terms.
Applied Questions: Require contextual reasoning or extrapolation.
Multiple Correct Answers: Include questions with more than one correct option.
Options and Answers:

Provide logical, clear, and unambiguous options.
Ensure the correct answer(s) can be justified based on the provided content or closely related concepts.
Explanation:

For every question, include a concise explanation that clarifies the answer using either the provided content or inferred knowledge.
Important Notes:
Use UPSC Benchmarks:
Adhere to the structure, depth, and logical reasoning typical of previous UPSC Prelims questions.

Beyond Content Where Necessary:

If the content lacks specific details but references important themes, infer and create questions based on your knowledge of the subject.
Avoid trivial or overly simplistic questions that fail to challenge the aspirant.
Avoid Assumptions:
Base questions and answers on sound reasoning, logical connections, or established facts, avoiding baseless assumptions.

Output Format:
For each question:

Question: [Insert question text here]
Options (if applicable): [Insert options here]
Answer: [Insert correct answer(s)]
Explanation: [Provide a concise explanation for the answer based on the content or inferred concepts]
Content:
<content starts here>
{content}
<content ends here>

'''
```

```python


def questions(prompt, client, model):
    messages = [
        {
            "role": "system",
            "content": "You are a helpful assistant that prepares the questions equivalent to UPSC Prelims standards"
        },
        {
            "role": "user",
            "content": prompt,
        }
    ]

    stream = client.chat.completions.create(
        model=model,
        messages=messages,
        max_tokens=8000,
        temperature=0.2,
        stream=False
    )
    return stream  
    
```

```python
result=questions(prompt_04_a, client, model)
result = result.choices[0].message.content
print(result)
```
