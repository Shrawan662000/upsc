```python
import fitz

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    extracted_text = ""

    # Iterate through all the pages
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        extracted_text += page.get_text()  # Extract text from the page
    
    pdf_document.close()
    return extracted_text

# Example usage
pdf_path = "01-06.pdf"  # Replace with your PDF file path
content = extract_text_from_pdf(pdf_path)
print(content)
```

```python
prompt=f'''
You are tasked with generating questions from the given content, adhering to the UPSC (Union Public Service Commission) Civil Services Prelims Examination standard. Analyze the content carefully and generate questions matching the depth and complexity of previous UPSC prelims questions. Each section of the content corresponds to a unique topic. If possible, generate at least one question per section based on the following guidelines.

Content:
{content}

Question Types to Generate:
Type-01: Multi-Statement Questions with Answer Codes
Example:
"Which of the following statements are correct about the Constitution of India?"

Powers of the Municipalities are given in Part IX A of the Constitution.
Emergency provisions are given in Part XVIII of the Constitution.
Provisions related to the amendment of the Constitution are given in Part XX of the Constitution.
Select the answer using the code given below:
a) 1 and 2 only
b) 2 and 3 only
c) 1 and 3 only
d) 1, 2 and 3
Answer: d) 1, 2 and 3
Explanation: All the statements correctly describe the respective parts of the Constitution of India.

Type-02: Single Statement Objective Questions
Example:
"Which one of the following statements is correct as per the Constitution of India?"

a) Inter-State trade and commerce is a State subject under the State List.
b) Inter-State migration is a State subject under the State List.
c) Inter-State quarantine is a Union subject under the Union List.
d) Corporation tax is a State subject under the State List.
Answer: c) Inter-State quarantine is a Union subject under the Union List.
Explanation: Inter-State quarantine is listed as a Union subject in the Constitution of India under the Union List.

Type-03: Statement Analysis Questions
Example:
"Consider the following statements in respect of the Constitution Day:"

Statement-I: The Constitution Day is celebrated on 26th November every year to promote constitutional values among citizens.
Statement-II: On 26th November 1949, the Constituent Assembly of India set up a Drafting Committee under the Chairmanship of Dr. B. R. Ambedkar to prepare a Draft Constitution of India.
Which one of the following is correct in respect of the above statements?

a) Both Statement-I and Statement-II are correct and Statement-II is the correct explanation for Statement-I
b) Both Statement-I and Statement-II are correct and Statement-II is not the correct explanation for Statement-I
c) Statement-I is correct but Statement-II is incorrect
d) Statement-I is incorrect but Statement-II is correct
Answer: c) Statement-I is correct but Statement-II is incorrect
Explanation: The Constitution Day is indeed celebrated on 26th November every year, but the Drafting Committee was formed on 29th August 1947, not 26th November 1949.

Type-04: Pair Matching Questions
Example:
"Match List I (Item in the Indian Constitution) with List II (Country from which it was derived):"

List I	                                                                 List II
A. Directive Principles of State Policy	                                1. Ireland
B. Fundamental Rights	                                                2. USA
C. Concurrent List in Union-State Relations	                            3. Australia
D. India as a Union of States with Greater Powers to Union	            4. Canada
Select the correct answer using the codes given below:

a) A-1, B-2, C-3, D-4
b) A-2, B-1, C-4, D-3
c) A-3, B-4, C-2, D-1
d) A-1, B-4, C-3, D-2
Answer: a) A-1, B-2, C-3, D-4

Explanation: Directive Principles were inspired by Ireland, Fundamental Rights by the USA, the Concurrent List by Australia, and the Union of States by Canada.

Note : Do not generate the any question forcefully, if it does not have relevant and enough information in content, and if also it does suits any type of question format.

Guidelines for Generating Questions:
Content-based: Ensure questions are strictly answerable using the given content. Do not include information outside the provided text.
Options and Answers: Decide the options and correct answers only from the given content without making assumptions.
UPSC Standard: Maintain the depth and complexity typical of UPSC prelims. Questions should be unambiguous and logical.

Instructions for Answer and Explanation:
Include the correct answer for each question.
Provide a concise explanation for the answer based on the content.


Desired Output Format: desired ouput format will be exactly like above given examples except question types.

Question: [Insert question text here]
Options (if applicable): [Insert options here]
Answer: [Correct answer(s)]
Explanation: [A concise explanation for the correct answer]

'''
```

```python


def questions(prompt, client, model):
    messages = [
        {
            "role": "system",
            "content": "You are a helpful assistant that prepares the questions equivalent to UPSC Prelims standards"
        },
        {
            "role": "user",
            "content": prompt,
        }
    ]

    stream = client.chat.completions.create(
        model=model,
        messages=messages,
        max_tokens=8000,
        temperature=0.5,
        stream=False
    )
    return stream  
    
```

```python
result=questions(prompt, client, model)
result = result.choices[0].message.content
print(result)
```
