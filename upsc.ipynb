```python
import fitz

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    extracted_text = ""

    # Iterate through all the pages
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        extracted_text += page.get_text()  # Extract text from the page
    
    pdf_document.close()
    return extracted_text

# Example usage
pdf_path = "data/01-06.pdf"  # Replace with your PDF file path
content = extract_text_from_pdf(pdf_path)
print(content)
```

```python
prompt=f'''You are tasked with generating questions from the below given content, adhering to the UPSC (Union Public Service Commission) civil service prelims examination standard. Keep in mind the previous year question of UPSC CSE prelims and generate the question matching that standard depth.
 Each section in the content corresponds to a unique topic. analyse the content first and then if possible generate at least one questions per section based on the following guidelines:
## Content : {content}

Question Types:
a. MCQ (Multiple Choice Question): A type of question where you are provided with four options, and only one option is correct.
b. MSQ (Multiple Select Question): A type of question where you are provided with four options, and one or more options may be correct.
c. Pair Matching Question: A type of question where two columns are provided. The left column contains at least four items, and the right column contains at least four corresponding items. The task is to match each item on the left with its correct counterpart on the right.

Note: Questions will only be generated if they meet the UPSC prelims standard. Non-compliant questions will be ignored.


## Some Important Instruction:
1. The questions should be framed in a way that they are answerable from the given content.
2. Do not generate any questions outside the given content.
3. Decide the Options and correct answers only based on the given content, without making assumptions or deciding answers independently.

## Instructions for Answer and Explanation:
1. Include the correct answer for each question.
2. Provide a short explanation for why the answer is correct.



## Desired output format
Output Format:
Section Name
Question Type (e.g., MCQ, MSQ, Pair Matching):
Question text...
Options (if applicable)...
Answer:
Correct answer(s).
Explanation: A concise explanation for the answer.

'''
```

```python


def questions(prompt, client, model):
    messages = [
        {
            "role": "system",
            "content": "You are a helpful assistant that prepares the questions equivalent to UPSC Prelims standards"
        },
        {
            "role": "user",
            "content": prompt,
        }
    ]

    stream = client.chat.completions.create(
        model=model,
        messages=messages,
        max_tokens=8000,
        temperature=0.5,
        stream=False
    )
    return stream  
    
```

```python
result=questions(prompt, client, model)
result = result.choices[0].message.content
print(result)
```
